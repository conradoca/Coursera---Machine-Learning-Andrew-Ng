## Load the data

```python
# Loading all data on a dictonary
data = loadmat('ex7data2.mat')

# Convert the data into a numpy array
X = data['X']

# m = number of training examples
# n = number of features
(m,n) = X.shape
```

```python
knn=Kmeans(3, 10)
```

```python
knn.fit(X)
```

## Plotting the result

```python
plt.scatter(X[:,0],X[:,1], marker='o', facecolors='none', c=knn.idCentroids, alpha=0.2)
plt.plot(knn.centroids[:, :,0],knn.centroids[:, :,1], marker='x', c='Black')
plt.scatter(knn.centroids[-1, :,0],knn.centroids[-1, :,1], 
         s=50, 
         facecolors='none', 
         edgecolors='Black', 
         alpha=1)
```
